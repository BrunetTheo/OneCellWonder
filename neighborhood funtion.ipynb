```python
import numpy as np

# 1. SETUP (Comme avant)
world_random = (np.random.rand(10, 10) < 0.35).astype(int)
neighborhood_counter = np.zeros_like(world_random) # Ta matrice de stockage

mask = np.array([
    [0, 1, 0],
    [1, 0, 1],
    [1, 1, 1]])

y = 9
x = 0

def neighborhood(world_random, y, x):
    rows, cols = world_random.shape
    
    y_start, y_end = max(0, y-1), min(rows, y+2)
    x_start, x_end = max(0, x-1), min(cols, x+2)
    
    world_slice = world_random[y_start:y_end, x_start:x_end]

    mask_slice = mask[
        (y_start - (y-1)) : (3 - ((y+2) - y_end)), 
        (x_start - (x-1)) : (3 - ((x+2) - x_end))
    ]
    
    # Calcul du nombre de voisins
    result = np.sum(world_slice * mask_slice)
    
    # --- CORRECTION ICI ---
    # On stocke le 'result' qu'on vient de calculer, on ne rappelle pas la fonction !
    neighborhood_counter[y, x] = result 
    
    return result 

# Test
print(f"Valeur calculée : {neighborhood(world_random, y, x)}")
print(f"Valeur stockée dans la matrice en ({y},{x}) : {neighborhood_counter[y, x]}")
```

```python
#exemple:
y = 9
x = 0


def neighborhood(world_random, y, x):
    #mask positionment arround the cell we are looking at
    rows, cols = world_random.shape
    
    #definition of the space limits so we don't try to look outside the world array
    y_start, y_end = max(0, y-1), min(rows, y+2)
    x_start, x_end = max(0, x-1), min(cols, x+2)
    #setting the max at 0 prevent us to have negative indexes so if we look at 0,1 we will look at 0,0 to 1,2
    #(so the cut will be smaller than 3x3)
    
    #Mask cuttinh
    world_slice = world_random[y_start:y_end, x_start:x_end]

    #If we are at the border we cut the mask    
    mask_slice = mask[
        (y_start - (y-1)) : (3 - ((y+2) - y_end)), 
        (x_start - (x-1)) : (3 - ((x+2) - x_end))
    ]
    
    #counter
    result = np.sum(world_slice * mask_slice)
    neighborhood_counter[y, x] = neighborhood(world_random, y, x)
    
    return result 
print(neighborhood(world_random, y, x))
```
